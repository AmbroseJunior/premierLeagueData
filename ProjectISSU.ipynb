CLUSTERING

Data Preparation

```python
performance_cols = ['Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt']
progression_cols = ['PrgC', 'PrgP', 'PrgR']
per_90_cols = ['Gls_90', 'Ast_90', 'G+A_90', 'G-PK_90', 'xG_90', 'npxG_90', 'xAG_90', 'npxG+xAG_90']

clustering_data = teams[performance_cols + progression_cols + per_90_cols]
```

Determine the Number of Clusters using the Elbow Method

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(clustering_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Apply K-Means; display & Analyse clusters

```python
kmeans = KMeans(n_clusters=3, random_state=42)
teams['cluster'] = kmeans.fit_predict(clustering_data)

for cluster in range(3):
    print(f"Cluster {cluster}:")
    cluster_data = teams[teams['cluster'] == cluster]
    print(f"Teams: {cluster_data['Team'].unique()}") # Display unique team names
    print(cluster_data[performance_cols + progression_cols + per_90_cols].mean())
    print("\n")
```

 Visualizing Clusters – Identify Traits and Patterns

Cluster 0 (Low Performers): This cluster might have significantly lower values across most metrics, indicating weaker attacking and overall performance.

Cluster 1 (High Performers): This cluster might stand out with the highest values in goals, assists, and related metrics, suggesting strong attacking capabilities.

Cluster 2 (Mid-Range Performers): This cluster might show moderate performance, falling between the other two clusters.

Uniform Manifold Approximation and Projection (UMAP)

```python
import pandas as pd
import matplotlib.pyplot as plt
import umap
import seaborn as sns

clustering_features = performance_cols + progression_cols + per_90_cols
X = teams[clustering_features]


reducer = umap.UMAP(n_components=2, random_state=42)
umap_components = reducer.fit_transform(X)

umap_df = pd.DataFrame(data=umap_components, columns=['UMAP1', 'UMAP2'], index=teams.index)
umap_df['cluster'] = teams['cluster']


plt.figure(figsize=(10, 6))
sns.scatterplot(x='UMAP1', y='UMAP2', hue='cluster', data=umap_df, palette='plasma', s=100)
plt.title('Team Visuals')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

The UMAP plot visualizes the relationships between Premier League teams based on their performance, progression, and per-90-minute metrics. Teams are grouped into three distinct clusters based on their similarities in these features.

Cluster 0 represents teams with lower performance and ball progression,

while Cluster 1 includes higher-performing teams with stronger attacking capabilities.

Cluster 2 consists of teams with moderate performance falling between the other two clusters.

The visualization highlights the separation of teams based on their overall playing styles and efficiency.

Manchester City appears as a potential outlier, suggesting a unique or specialized approach to player efficiency and productivity compared to other teams.

This UMAP analysis provides insights into the tactical and strategic differences among Premier League teams.
